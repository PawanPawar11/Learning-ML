```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
years_of_experience = 7
```

```python
raw_data = {
    'years_of_experience_of_employee': [1, 2, 3, 4, 6, 8, 11, 13, 16, 20],
    'salary_based_on_it': [4, 5, 7, 12, 18, 20, 25, 30, 40, 55]
}
```

```python
print(raw_data)
```

```python
df = pd.DataFrame(raw_data)
```

```python
print(df)
```

```python
X = np.array(df['years_of_experience_of_employee']).reshape(-1, 1)
# what does negative values mean here, 
# does it mean that automatically infer the amount of columns we'll need,
# if the amount of values in each column will be exactly euqal to 1
```

```python
y = np.array(df['salary_based_on_it']).reshape(-1, 1)
```

```python
print(X)
```

```python
print(y)
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, random_state = 42, test_size = 0.20)
```

```python
print(train_X)
```

```python
print(test_X)
```

```python
print(train_y)
```

```python
print(test_y)
```

```python
model = LinearRegression()
```

```python
model.fit(train_X, train_y)
```

```python
y_prediction = model.predict([[years_of_experience]])
```

```python
print("PREDICTION:", y_prediction)
```

```python
y_test_prediction = model.predict(test_X)
```

```python
print(y_test_prediction)
```

```python
y_line = model.predict(X)
```

```python
print(y_line)
```

```python
print("SLOPE:", model.coef_)
print("-----------------------")
print("INTERCEPT:", model.intercept_)
```

```python
print("MAE:", mean_absolute_error(test_y, y_test_prediction))
# what will happen if we alter the order of arguments?
```

```python
print("r2:", r2_score(test_y, y_test_prediction))
```

```python
plt.scatter(X, y, s = 12)
plt.xlabel("Year of Experience")
plt.ylabel("Salary")
plt.plot(X, y_line, color = "r")
plt.show()
```
